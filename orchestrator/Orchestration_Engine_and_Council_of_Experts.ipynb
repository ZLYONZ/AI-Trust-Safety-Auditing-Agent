```python
def normalize_results(results):

    normalized = {}

    for name, r in results.items():
        normalized[name] = {
            "score": r.module_score,
            "risk": r.risk_level,
            "findings": r.findings
        }

    return normalized
```

```python
class AuditOrchestrator:

    def __init__(self):
        from governance_module import GovernanceModule
        from fairness_module import FairnessModule
        from security_module import SecurityModule
        from transparency_module import TransparencyModule
        from performance_module import PerformanceModule

        self.governance = GovernanceModule()
        self.fairness = FairnessModule()
        self.security = SecurityModule()
        self.transparency = TransparencyModule()
        self.performance = PerformanceModule()

    def run(self, document_text):

        print(" Running AI Audit Pipeline...\n")

        results = {}

        # -----------------------------
        # Run Each Module
        # -----------------------------
        print("Running Governance Module...")
        results["governance"] = self.governance.run(document_text)

        print("Running Fairness Module...")
        results["fairness"] = self.fairness.run(document_text)

        print("Running Security Module...")
        results["security"] = self.security.run(document_text)

        print("Running Transparency Module...")
        results["transparency"] = self.transparency.run(document_text)

        print("Running Performance Module...")
        results["performance"] = self.performance.run(document_text)

        normalized_results = normalize_results(results)

        # -----------------------------
        # Build Summary
        # -----------------------------
        summary = {}

        for name, res in results.items():
            summary[name] = {
                "score": res.module_score,
                "risk": res.risk_level,
                "severity": "PASS" if res.module_score >= 0.75 else "FAIL"
            }

        return {
            "results": normalized_results,
            "summary": summary
        }
```

# **Council of Expert**

```python
def generate_peer_reviews(module_results):
    """
    Generate cross-module peer reviews (Council of Experts - Stage 2)
    """

    reviews = []

    modules = list(module_results.keys())

    for reviewer in modules:
        for target in modules:

            if reviewer == target:
                continue

            reviewer_score = module_results[reviewer]["score"]
            target_score = module_results[target]["score"]

            score_gap = abs(reviewer_score - target_score)

            # -----------------------------
            # Cross-domain score (1-10)
            # -----------------------------
            cross_score = max(1, int(10 - score_gap * 10))

            # -----------------------------
            # Consistency Check
            # -----------------------------
            if score_gap < 0.1:
                consistency = "consistent"
            elif score_gap < 0.25:
                consistency = "partial"
            else:
                consistency = "inconsistent"

            # -----------------------------
            # Blind-spot detection
            # -----------------------------
            issues = []

            if reviewer == "governance" and target == "fairness":
                if target_score > 0.8 and reviewer_score < 0.8:
                    issues.append("Fairness appears strong but governance framework may not enforce bias accountability")

            if reviewer == "performance" and target == "fairness":
                if target_score > 0.8:
                    issues.append("Fairness metrics should be validated with model performance across demographics")

            if reviewer == "security" and target == "performance":
                if target_score > 0.9:
                    issues.append("High performance does not guarantee robustness or adversarial security")

            if reviewer == "transparency" and target == "governance":
                if reviewer_score < 0.75:
                    issues.append("Governance may lack sufficient explainability and disclosure mechanisms")

            if score_gap > 0.3:
                issues.append("Significant score divergence between modules")

            if len(issues) == 0:
                issues.append("No major issues identified")

            # -----------------------------
            # Recommendation
            # -----------------------------
            if score_gap < 0.1:
                recommendation = "agree"

            elif score_gap < 0.25:
                recommendation = "flag_for_review"

            else:
                recommendation = "escalate"

            # -----------------------------
            # Build Review Card
            # -----------------------------
            review_card = {
                "reviewer": reviewer,
                "target": target,
                "cross_score": cross_score,
                "consistency": consistency,
                "issues": issues,
                "recommendation": recommendation
            }

            reviews.append(review_card)

    return reviews
```

```python
peer_reviews = generate_peer_reviews(final_result["results"])

for r in peer_reviews[:5]:
    print(r)
```

```python
def summarize_peer_reviews(peer_reviews):

    total = len(peer_reviews)

    agree = sum(1 for r in peer_reviews if r["recommendation"] == "agree")
    flag = sum(1 for r in peer_reviews if r["recommendation"] == "flag_for_review")

    inconsistent = sum(1 for r in peer_reviews if r["consistency"] == "inconsistent")

    return {
        "total_reviews": total,
        "agree": agree,
        "flag_for_review": flag,
        "inconsistent_cases": inconsistent
    }

peer_summary = summarize_peer_reviews(peer_reviews)
print(peer_summary)
```

```python
def compute_divergence(scores):
    spread = max(scores) - min(scores)

    if spread >= 0.35:
        return "critical"
    elif spread >= 0.2:
        return "high"
    elif spread >= 0.1:
        return "moderate"
    else:
        return "low"

scores = [info["score"] for info in final_result["summary"].values()]
divergence = compute_divergence(scores)
print(divergence)
```

```python
def arbitrator_synthesis(scores, divergence, peer_summary):
    """
    Final decision engine (Stage 3)

    Inputs:
        scores: list of module scores (0~1)
        divergence: "low" / "moderate" / "high" / "critical"
        peer_summary: summarize_peer_reviews() output

    Returns:
        dict with final decision
    """

    # -----------------------------
    # 1. Convert score → 0-100 scale
    # -----------------------------
    scores_100 = [s * 100 for s in scores]
    avg_score = sum(scores_100) / len(scores_100)

    # -----------------------------
    # 2. Confidence calculation
    # -----------------------------
    agree_ratio = peer_summary["agree"] / peer_summary["total_reviews"]
    inconsistency_ratio = peer_summary["inconsistent_cases"] / peer_summary["total_reviews"]

    confidence = round(
        0.6 * agree_ratio + 0.4 * (1 - inconsistency_ratio),
        3
    )

    # -----------------------------
    # 3. Decision logic
    # -----------------------------
    if divergence == "critical":
        decision = "ESCALATE"
        notes = "Critical divergence across modules → mandatory human review"

    elif divergence == "high":
        decision = "ESCALATE"
        notes = "High disagreement detected → human-in-the-loop required"

    elif divergence == "moderate":
        decision = "REVIEW"
        notes = "Moderate disagreement → include review notes"

    else:
        # divergence low → see overall score
        if avg_score >= 75:
            decision = "APPROVE"
            notes = "Strong performance with high agreement"

        elif avg_score >= 60:
            decision = "REVIEW"
            notes = "Acceptable but needs minor review"

        else:
            decision = "ESCALATE"
            notes = "Low overall score despite low divergence"

    # -----------------------------
    # 4. Final Output
    # -----------------------------
    return {
        "final_score": round(avg_score, 2),
        "confidence": confidence,
        "decision": decision,
        "divergence": divergence,
        "notes": notes
    }
```

```python
final_decision = arbitrator_synthesis(scores, divergence, peer_summary)

print("\n===== FINAL DECISION =====")
print(final_decision)
```
